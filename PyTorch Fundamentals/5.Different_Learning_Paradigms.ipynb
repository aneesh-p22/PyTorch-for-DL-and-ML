Types of Learning

Supervised Learning
- When you have a lot of data (inputs) and labels (what that input should look like at the output)
- e.g. Labelled dog and cat images

Unsupervised and Self-supervised learning
- When you just have data and no labels
- The model can figure out the fundamental pattern between the data but doesn't know what it actually is
- e.g. unlabelled dog and cat images
    - The model can identify the patterns between the two, but wouldn't be able to say "This is a dog" or "This is a cat", it could only say that they are two different types of things
- You can inform it of the meanings of the patterns later

Transfer Learning
- Taking the patterns that one model has learned from a dataset and transferring it to another model
- Gives model a headstart, can save time and resources

Reinforcement Learning
- You have an environment and in that environment, an agent that makes decisions through trial and error
- You give rewards and observations back to that agent
- The agent adjusts its behaviour over time to maximise cumulative rewards
