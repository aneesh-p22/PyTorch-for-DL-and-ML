Deep Learning Use Cases

- Sequence to Sequence (seq2seq)
    - You put one sequence in and get one sequence out
    - Examples:
        - Translation
        - Speech recognition

- Classification / Regression
    - Regression predicts a number for detection, classification identifies what is detected (predicting what something is)
    - Examples:
        - Computer vision
        - Natural Language Processing (NLP)

- Other examples:
    - Recommendation systems
    - etc...
