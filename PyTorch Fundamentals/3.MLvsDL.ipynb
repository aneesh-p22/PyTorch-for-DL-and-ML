Typically they can be used interchangeably

DL is a subset of ML

Traditional ML algorithms are used on structured data, such as tables of numbers
- e.g. XGBoost

DL algorithms are better for unstructured data, such as natural language, images, audio data
- e.g. Neural networks

Common ML algorithms (shallow algorithms):
- Random Forest
- Gradient boosted models
- Naive Bayes
- Nearest neighbour
- Support vector machine

Common DL algorithms
- Neural networks
- Fully connected neural network
- Convolutional neural network
- Recurrent neural network
- Transformers

Depending on the problem, many of the algorithms can be used for both ML and DL
