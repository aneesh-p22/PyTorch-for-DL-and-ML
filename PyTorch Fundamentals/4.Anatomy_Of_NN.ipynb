Inputs
- We have some unstructured data (image, words, audio)

Numerical Encoding
- Before data can be used in an NN, it needs to be turned into numbers

Learns representation (patterns/features/weights)
- These are then passed into a neural network which has layers:
    - Input layer
    - Hidden layers
    - Output layer
- Each layer has nodes/neurons
- Different NNs used for different problems / inputs
- The NN manipulates the data and learns representations (patterns in the data) that it thinks best represents the data (on their own, we don't define anything)

Outputs
- Outputs the representation outputs
    - Features/weight matrix/weight tensor/learned representation
- Outputs are converted to human understandable outputs

Anatomy

Input layer
- Data goes in
- Has some number of units/nodes/neurons

Hidden layer(s)
- Learns patterns in data
    - Patterns is an arbitrary term, embeddings/weights/feature representation/feature vectors are all similar things
- Has some number of neurons and layers
- Deep in DL comes from having lots of layers

Output layer
- Outputs learned representation or prediction probabilities

Each layer is usually a combination of linear (straight line) and/or non-linear (not-straight line) functions
- A NN uses this to draw patterns from the data
