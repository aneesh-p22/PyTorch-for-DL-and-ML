What DL is good for:
- Problems with long lists of rules
    - When the traditional approach fails, ML/DL may help
- Continually changing environments
    - DL can adapt ('learn') to new scenarios
- Discovering insights within large collections of data
    - Imagine trying to hand-craft rules for a huge complex set of data

What DL is (typically) not good for:
- When you need explainability
    - The patterns learned by a deep learning model are typically uninterpretable by a human
    - Patterns learned by ML models are lots of numbers called weights and biases which are typically uninterpretable by humans
    - Models can have very large numbers of parameters (numbers or patterns in data since ML is turning things into numbers and then writing an ML model to find patterns in those numbers)
- When the traditional approach is a better option
    - If you can accomplish what you need with a simple rule-based system
- When errors are unacceptable
    - Since the outputs of deep learning models aren't always predictable
    - DL models make a probabilistic bet on what the outcome is and despite the model being confident, it may not always be correct
    - In a rule-based system you know what the outputs will be every time
- When you don't have much data
    - DL models require a fairly large amount of data to produce great results
